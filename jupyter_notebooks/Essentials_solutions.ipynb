```python
import numpy as np
import matplotlib.pyplot as plt
```

# Question 1

```python
def f(x):
    return np.exp(-4*np.sin(2*np.pi*x))
```

```python
x = np.linspace(0, 1, 1000)
y = f(x)
```

```python
plt.plot(x,y)
```

# Question 2

## Part A

```python
dx = x[1] - x[0] 
integral = np.sum(y*dx)
integral
```

## Part B

```python
cumulative_integral = np.cumsum(y*dx)
```

```python
plt.plot(x, cumulative_integral)
```

# Question 3

## Part B

```python
dx = np.diff(x)
dfdx = df/dx
```
